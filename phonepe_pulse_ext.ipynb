```python
import os
data_list = []
# Base path to the directory containing JSON files
base_directory = r"D:\git_pro\pulse\data\aggregated\transaction\country\india\state"

# Iterate through states
for state_folder in os.listdir(base_directory):
    state_path = os.path.join(base_directory, state_folder)
#     print(state)
    if os.path.isdir(state_path):
        # Iterate through years
        for year_folder in os.listdir(state_path):
            year_path = os.path.join(state_path, year_folder)
#             print(year_folder)
            if os.path.isdir(year_path):
                # Iterate through JSON files
                for json_file in os.listdir(year_path):
                    if json_file.endswith('.json'):
                        json_file_path = os.path.join(year_path, json_file)
#                         print(json_file)
                        with open(json_file_path, 'r') as file:
                            data = json.load(file)
                            
                            # Extract the data we're interested in
                            for transaction_data in data['data']['transactionData']:
                                row_dict = {
                                    'States': state_folder,
                                    'Transaction_Year': year_folder,
                                    'Quarters': int(json_file.split('.')[0]),
                                    'Transaction_Type': transaction_data['name'],
                                    'Transaction_Count': transaction_data['paymentInstruments'][0]['count'],
                                    'Transaction_Amount': transaction_data['paymentInstruments'][0]['amount']
                                }
                                data_list.append(row_dict)
```

```python
import os
import json

data_list2 = []
base_directory2 = r"D:\git_pro\pulse\data\aggregated\user\country\india\state"

for state_folder in os.listdir(base_directory2):
    state_path = os.path.join(base_directory2, state_folder)
    if os.path.isdir(state_path):
        for year_folder in os.listdir(state_path):
            year_path = os.path.join(state_path, year_folder)
            if os.path.isdir(year_path):
                aggregated_data = None  # Initialize aggregated_data variable
                for json_file in os.listdir(year_path):
                    if json_file.endswith('.json'):
                        json_file_path = os.path.join(year_path, json_file)
                        with open(json_file_path, 'r') as file:
                            json_data = json.load(file)
                            if 'aggregated' in json_data['data']:
                                aggregated_data = json_data['data']['aggregated']
                        
                if aggregated_data:
                    row_dict = {
                        'States': state_folder,
                        'Transaction_Year': year_folder,
                        'Register_users': aggregated_data['registeredUsers'],
                        'App_opens': aggregated_data['appOpens']
                    }
                    data_list2.append(row_dict)
```

```python
import os
import json

data_list6 = []
data_list7 = []
base_directory6 = r"D:\git_pro\pulse\data\top\user\country\india\state"

for state_folder in os.listdir(base_directory6):
    state_path = os.path.join(base_directory6, state_folder)
    if os.path.isdir(state_path):
        for year_folder in os.listdir(state_path):
            year_path = os.path.join(state_path, year_folder)
            if os.path.isdir(year_path):
                for json_file in os.listdir(year_path):
                    if json_file.endswith('.json'):
                        json_file_path = os.path.join(year_path, json_file)
                        with open(json_file_path, 'r') as file:
                            data = json.load(file)
                            
                            for district in data['data']['districts']:
                                row_dict = {
                                    'States': state_folder,
                                    'Transaction_Year': year_folder,
                                    'Quarters': int(json_file.split('.')[0]),
                                    'District': district['name'],
                                    'RegisteredUsers': district['registeredUsers']
                                }
                            data_list6.append(row_dict)
                            
                            for pincode in data['data']['pincodes']:
                                row_dict = {
                                    'States': state_folder,
                                    'Transaction_Year': year_folder,
                                    'Quarters': int(json_file.split('.')[0]),
                                    'District_pin_code': pincode['name'],
                                    'RegisteredUsers_bypin_code': pincode['registeredUsers']
                                }
                            data_list7.append(row_dict)
```
